```python
import re
import pandas as pd

path = 'C:/Users/shree/OneDrive/Desktop/UserCase/New/AltaML_FHX_SAMA_Data_Samples/Reduced_FHX_file.txt'
rawdata= open(path,'r')
data = rawdata.read()
#print(data)

#Module Class - Main Module
MC_name=[]

#Module_Instance_Name
MI_name=[]

#Function Class and Defination
fb_names =[]
defi=[]

#Wire Source and Destination
ws =[]
des=[]

pattern = re.compile(r"MODULE_CLASS NAME=\W(.+)\W\sC+")
match_MC = pattern.finditer(data)
for i in match_MC:
    MC_name.append(i.group(1))
Module_Class = pd.DataFrame(MC_name,columns=["Module_Class"])
#Module_Class

pattern1 = re.compile(r"MODULE_INSTANCE TAG=\W(.+)\W\sPL")
match_MI = pattern1.finditer(data)
for i in match_MI:
    MI_name.append(i.group(1))
Module_Instance=pd.DataFrame(MI_name,columns=["Module_Instance"])
#Module_Instance

pattern2 = re.compile(r"FUNCTION_BLOCK NAME=\W(.+)\W\sDEFINITION=\W(.+)\W")
match_FB = pattern2.finditer(data)
for i in match_FB:
    fb_names.append(i.group(1))
    defi.append(i.group(2))

FC = pd.DataFrame(columns = ['Function_Block_Name','Defination'])
FC["Function_Block_Name"]=fb_names
FC["Defination"]=defi
#FC

pattern = re.compile(r"WIRE SOURCE=\W(.+)\W DESTINATION=\W(.+)\W")
match_WSD = pattern.finditer(data)
for i in match_WSD:
    ws.append(i.group(1))
    des.append(i.group(2))
    
WS = pd.DataFrame(columns = ['Wire_Source','Destination'])
WS["Wire_Source"]=ws
WS["Destination"]=des
#WS
```
